#1. Análisis de facturación



En este primer análisis, debes calcular el ingreso total de cada tienda. **Sumando los valores de la columna Precio** de cada conjunto de datos de la tienda para estimar los ingresos.

```python
print(sum(tienda.Precio))
print(sum(tienda2.Precio))
print(sum(tienda3.Precio))
print(sum(tienda4.Precio))
```

```python
def total_facturado(tienda) -> int:
  suma_total = int(sum(tienda))
  return int(suma_total)

facturacion_t1 = total_facturado(tienda.Precio)
print(f"Los ingresos estimados por la tienda 1 fueron: {facturacion_t1}")
facturacion_t2 = total_facturado(tienda2.Precio)
print(f"Los ingresos estimados por la tienda 2 fueron: {facturacion_t2}")
facturacion_t3 = total_facturado(tienda3.Precio)
print(f"Los ingresos estimados por la tienda 3 fueron: {facturacion_t3}")
facturacion_t4 = total_facturado(tienda4.Precio)
print(f"Los ingresos estimados por la tienda 4 fueron: {facturacion_t4}")
```

```python
nombre_tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
facturacion = [total_facturado(tienda.Precio), total_facturado(tienda2.Precio), total_facturado(tienda3.Precio), total_facturado(tienda4.Precio)]

print(nombre_tiendas, facturacion)
```

```python
facturacion_tienda = list(zip(nombre_tiendas, facturacion))
facturacion_tienda
```

```python
data_facturacion = {'Tiendas': nombre_tiendas,
                    'Facturacion': facturacion}

df_facturacion = pd.DataFrame(data_facturacion)
df_facturacion.head()
```

```python
def mostrar_valores(pct, valores):
    total = sum(valores)
    valor_absoluto = int(round(pct * total / 100.0))
    return f"${valor_absoluto:,.0f}".replace(',', '.')

explode = [0.05] * len(df_facturacion)

plt.figure(figsize=(10,7))
plt.pie(df_facturacion['Facturacion'],
        labels = df_facturacion['Tiendas'].values,
        explode=explode,
        autopct=lambda pct: mostrar_valores(pct, facturacion),
        textprops= {'size' : 'x-large',
            'fontweight' : 'bold',
            'color' : 'w'})
plt.legend()
plt.axis('equal')
plt.title('Facturacion total por tienda', fontsize = '18', fontweight = 'bold')
plt.show()
```

```python
#print(f"La tienda que generó mas ventas fue: {max(facturacion)}")
#print(f"La tienda que generó menos ventas fue: {min(facturacion)}")

print(f"La tienda que generó mas ventas fue: {df_facturacion['Facturacion'].max()}")
print(f"La tienda que generó menos ventas fue: {df_facturacion['Facturacion'].min()}")
```

# 2. Ventas por categoría

En este paso, vas a agrupar los datos por categoría y contar cuántas ventas hubo de cada tipo. Esto te permitirá identificar las categorías más populares por tienda.

```python
print(tienda['Categoría del Producto'])

def obtener_categoria(tienda):
  categoria_producto = tienda
  return categoria_producto

tienda1_categoria = obtener_categoria(tienda['Categoría del Producto'])
tienda2_categoria = obtener_categoria(tienda2['Categoría del Producto'])
tienda3_categoria = obtener_categoria(tienda3['Categoría del Producto'])
tienda4_categoria = obtener_categoria(tienda4['Categoría del Producto'])
```

```python
def contador_categorias(tienda):
  contador = tienda.value_counts()
  return contador

listado_cat_t1 = contador_categorias(tienda1_categoria)
listado_cat_t2 = contador_categorias(tienda2_categoria)
listado_cat_t3 = contador_categorias(tienda3_categoria)
listado_cat_t4 = contador_categorias(tienda4_categoria)
```

```python
listado_cat_tiendas = [listado_cat_t1, listado_cat_t2, listado_cat_t3, listado_cat_t4]
listado_cat_tiendas
```

```python
tienda1_categorias_unicas = tienda['Categoría del Producto'].unique()

df_contador_categorias = listado_cat_t1.reset_index()
df_contador_categorias.columns = ['Categoria del Producto', 'Cantidad de Ventas']


def imprime_grafico(df_contador_categorias, titulo):
  plt.figure(figsize=(10,7))

  ax = sns.barplot(x = 'Cantidad de Ventas', y = 'Categoria del Producto', data = df_contador_categorias, edgecolor="0.2",
                  palette=['skyblue', 'red', 'yellow', 'green', 'purple', 'gray'])
  for container in ax.containers:
    ax.bar_label(container, fmt='%d')
  plt.title(titulo, fontsize = '18', fontweight = 'bold')
  plt.show()

imprime_grafico(df_contador_categorias, "Tienda 1")
```

# 3. Calificación promedio de la tienda


En este paso del challenge, el objetivo es calcular el promedio de las evaluaciones para entender qué tan satisfechos están los clientes.

```python
def obtener_categoria(tienda):
  categoria_producto = tienda
  return categoria_producto

calificacion_tienda1 = obtener_categoria(tienda['Calificación'])
calificacion_tienda2 = obtener_categoria(tienda2['Calificación'])
calificacion_tienda3 = obtener_categoria(tienda3['Calificación'])
calificacion_tienda4 = obtener_categoria(tienda4['Calificación'])
```

```python
def calcula_promedio(lista) -> float:
  resultado = sum(lista)/len(lista)
  return resultado

promedio_tienda1 = round(calcula_promedio(calificacion_tienda1),2)
promedio_tienda2 = round(calcula_promedio(calificacion_tienda2),2)
promedio_tienda3 = round(calcula_promedio(calificacion_tienda3),2)
promedio_tienda4 = round(calcula_promedio(calificacion_tienda4),2)
```

# Extra. Ventas por localización

```python
def ventas_por_ubicacion(Dataframe):
  ventas_ubicacion = Dataframe.groupby(['lat', 'lon'])['Producto'].count().reset_index()
  ventas_ubicacion.rename(columns={'Producto': 'Total de ventas'}, inplace=True)
  return ventas_ubicacion

ventas_odenadas_t1 = ventas_por_ubicacion(tienda).sort_values(by='Total de ventas', ascending=False)

ventas_odenadas_t1.head()
```

```python
def imprime_mapa(Dataframe, NumeroTienda):
  fig = px.scatter_map(Dataframe,
                      lat="lat",
                      lon="lon",
                      color="Total de ventas",
                      title = NumeroTienda,
                      color_continuous_scale='Picnic',
                      size_max=15)
  fig.update_geos(
      visible=False,
      lonaxis=dict(range=[-120, -30]),
      lataxis=dict(range=[-60, 30]),
    )
  fig.show()

imprime_mapa(ventas_odenadas_t1, "Tienda 1 - Ventas por sector.")
```

```python
ventas_ordenadas_t2 = ventas_por_ubicacion(tienda2).sort_values(by='Total de ventas', ascending=False)
ventas_ordenadas_t2.head()

imprime_mapa(ventas_ordenadas_t2, "Tienda 2 - Ventas por sector.")
```

```python
ventas_ordenadas_t3 = ventas_por_ubicacion(tienda3).sort_values(by='Total de ventas', ascending=False)
ventas_ordenadas_t3.head()

imprime_mapa(ventas_ordenadas_t3, "Tienda 3 - Ventas por sector.")
```

```python
ventas_ordenadas_t4 = ventas_por_ubicacion(tienda4).sort_values(by='Total de ventas', ascending=False)
ventas_ordenadas_t4.head()

imprime_mapa(ventas_ordenadas_t4, "Tienda 4 - Ventas por sector.")
```
